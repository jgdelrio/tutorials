# Introduction to Recommender Systems

Recommender systems are very popular applications today and they are used in many sectors to personalize the service provided to customers. The basic concept of what this models do is to predict the "rating" or "preference" that a user would give to an item.

Almost every major tech company has applied recommender systems in some form. Amazon suggests products to customers, YouTube selects which video to play next on autoplay and Facebook recommend pages to like and people to follow. For companies like Netflix and Spotify it's the core of the business model and its success revolves around the power of their recommendations.


## Types of recommenders:

1. **Simple recommenders** offer generalized recommendations based on popularity and/or genre. The basic idea behind this system is that items that are more popular and acclaimed will have higher probability of being liked by the average audience. IMDB Top 250 is an example of this system.
2. **Content recommenders** suggest similar items based on a particular item. This system uses item metadata, for example in films they will use genre, director, description, actors, synopsis, etc. in music, the singer, writer, genre, etc, to make these recommendations. The general idea behind these recommender systems is that if someone liked a particular item the same person will also like an item that is similar to it.
3. **Collaborative filtering (CF) recommenders** try to predict the rating or preference that a user would give an item-based on past ratings and preferences of other users. Collaborative filters do not require item metadata like its content-based counterparts.
4. **Hybrid recommenders** can be created by mixing properties of two or more types of recommendation systems.

<br><br>
We can extend as well the classification of $\color{RoyalBlue}{\text{Collaborative filtering recommenders}}$:
* **Memory Based**: Easy creation and explanaition of results but with performance issues due to the sparse data and the scalability issues. In this models we don’t use parametric ML.
  * $\color{RoyalBlue}{\textit{User-item}}$ filtering: find users that are similar to that user based on similarity of ratings, and recommend items that those similar users liked.
  * $\color{RoyalBlue}{\textit{Item-item}}$ filtering: find users who liked that item and find other items that those users or similar users also liked.
  * $\color{RoyalBlue}{\textit{Others}}$: ex KNN (clustering) as the previous ones it doesn't try to optimize a parameter. Using similarities between users and/or items as weights to predict a rating for a user and an item. The difference with Pearson or Cosine is that the similarities in this approach are calculated based on an unsupervised learning model. In this approach, we also limit the number of similar users as k, which makes system more scalable.
  
* **Model Based**: Among its advantages, it contain features like dimentionality reduction that deals with missing or sparse data but inference and explaining the behaviour of the model is difficult due to the hidden/latent factors.
  * $\color{RoyalBlue}{\textit{Matrix Factorization}}$: the idea is that the preferences of a user can be determined by a small number of hidden factors. We can call these factors as Embeddings. The matrix decomposition can be reformulated as an optimization problem with loss functions and constraints. Now the constraints are chosen based on property of our model. For example for Non negative matrix decomposition, we want non negative elements in resultant matrices.
      * $\color{BlueViolet}{\text{SVD}}$
      * $\color{BlueViolet}{\text{Probabilistic Matrix Factorization}}$
      * $\color{BlueViolet}{\text{Non veMatrix Factorization}}$
  * $\color{RoyalBlue}{\textit{Deep Learning}}$
      * $\color{BlueViolet}{\text{Multi layered Neural Networks}}$

To clarify more about **embeddings**, we can understand them intuitively as low dimensional hidden factors for items and users. Let's say that we have 3 dimensional embeddings for both items and users (chosen randomly). Then for a given user and item we can say the those 3 numbers (embeddings) represent 3 different characteristics about the item (those characteristics can be anything, how old is the item, which one was its author, how big or well described is the item... etc). Likewise 3 numbers in user embedding matrix might represent any 3 characteristics regarding the user (teste, age, location, etc).

This those premises, a higher number from dot product of user-X-embeddings and item-Y-embeddings means that item-Y is a good recommendation for user-X. And no, we don't have any idea what those factors are, it's just the intuition of the model based on the data).

<br><br>
Common steps across the recommenders include:
1. Decide on the **metric or score** on which to rate the items.
2. **Calculate scores** for the existing items.
3. **Sort the items** based on the score and output the top results.

## Most Popular

For this calculation the most evident parameters in which to base the scoring are the movie ratings or averaging the user ratings. In both cases there are a few caveats. 

For example, it does not take into consideration the number of contributions to scoring a particular item so a movie with a rating of 5 (user rating is from 0 to 5 here) from 20 voters will be considered better than a movie with a rating of 8.5 from 1000 voters.

Basically it tends to favor items with smaller number of voters and extremely high ratings. As the number of voters increase the rating regularizes and the final score becomes more representative of the true quality or value of the item.

To compensate for that we can create a weighted rating that penalizes the rating of items with very few votes. Following the IMDB formula the resulting expression is:

\begin{equation*}
Weighted​​Rating​​(WR)​​​=​​​\left( ​{\frac{v}{v + m}} ​ AR \right)​+​\left( ​{\frac{m}{v + m}} ​ MV \right)
\end{equation*}

Parameters:
- v is the number of votes for the movie
- m is the minimum votes required to be listed in the chart
- AR is the average rating of the movie
- MV is the mean vote across the whole report

<br><br>
And now we can begin to calculate these parameters from the data available.

First for the Mean Vote across the full data we can just apply the mean in the dataframe:

```python
MV = metadata['vote_average'].mean()
print(f"Mean Vote: {MV:.3f}")
```

To calculate the min amount of votes we can consider items within a specific percentile. For example 90th or 95th and this way we will only consider items that have equal or more votes than 90% (or 95%) of the items.

```python
m = metadata['vote_count'].quantile(0.90)
print(f"90th percentile cut off: {int(m)}")

m = metadata['vote_count'].quantile(0.95)
print(f"95th percentile cut off: {int(m)}")
```

```python
plt.figure(figsize=(15, 8))
ax = metadata['vote_count'].hist(bins=100, range=[0, 400])
ax.set_title("Vote Count", weight='bold');
ax.set_xlabel("# Movies", labelpad=20, weight='bold', size=12);
ax.set_ylabel("Count", labelpad=20, weight='bold', size=12);
```

As expected the 95th percentile will be more restrictive requiring items with at least 434 votes.

Now we filter the movies that comply with the requirement of the 95th percentile:

```python
selected_movies = metadata.copy().loc[metadata['vote_count'] >= m]
selected_movies.shape
```

From these 2274 movies we calculate the metric for each one by using the Weighted Rating that we have previously defined.

```python
def weighted_rating(x, m=m, mean_vote=MV):
    n_votes = x['vote_count']
    vote_av = x['vote_average']
    # Calculation based on the IMDB formula
    return (n_votes/(n_votes + m) * vote_av) + (m/(m + n_votes) * mean_vote)
```

```python
# Apply Weighted Rating
selected_movies['score'] = selected_movies.apply(weighted_rating, axis=1)

# Sort movies based on the score
selected_movies = selected_movies.sort_values('score', ascending=False)

# Print the top 10 movies
TOPN = 10
rec_most_popular = selected_movies[['title', 'vote_count', 'vote_average', 'score']].head(500)
rec_most_popular.columns = ['itemId', 'vote_count', 'vote_average', 'score']
rec_most_popular[:TOPN]
```

## Content Recommender

This model builds recommendations based on similarity across items. More specifically we have to compute pairwise similarity scores for all items based on specific parameters like plot descriptions. In this dataset we have available the plot description.

This data includes a lot of words that are not usefull for the comparison, like stopwords, so we need to clean it first and then translate it to word vectors which is what we will use for the similarity calculations.

**Term Frequency-Inverse Document Frequency (TF-IDF)** calculated for each document will provide as with a matrix were each document is a row and each column is a different word (that must appears in at least one document). The TF-IDF score is the frequency of a word occurring in a document, down-weighted by the number of documents in which it occurs. This is done to reduce the importance of words that occur frequently in plot overviews and therefore, their significance in computing the final similarity score.

Fortunately, scikit-learn gives you a built-in TfIdfVectorizer class that produces the TF-IDF matrix in a couple of lines.

```python
# Example of the data...
metadata['overview'].head(3)
```

```python
# Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

# Define a TF-IDF Vectorizer Object and remove all english stop words such as 'the', 'a'
tfidf = TfidfVectorizer(stop_words='english')

# Replace NaN with an empty string
metadata['overview'] = metadata['overview'].fillna('')

# Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(metadata['overview'])
tfidf_matrix.shape
```

As a result over 75K different words were used to describe the 45K movies in the dataset.

Now we proceed to compute the similarity score. This score is basically calculating the distance between two vectors so be can find the pairs of vectors in which this distance is the smaller (so they are as similar as possible). By distance, we mean the angular distance between two vectors, which is represented by θ (theta). By thinking further from the machine learning perspective, we see that the value of **cos θ** makes more sense to us rather than the value of **θ (theta)** because, the cosine function will give us a value between 0 to 1 (mapping the value of θ in the first quadrant).


Typical similarity metrics include:
- **euclidean**
- **Pearson**
- **cosine**

Different scores work well in different scenarios and it is often a good idea to experiment with different metrics.

For example using the cosine similarity score which it is independent of magnitude and is relatively easy and fast to calculate we apply the following expression:


\begin{equation*}
cosine \left( x, y \right)​​=​​{\frac{x​\cdot​y^T}{||x||​\cdot​||y||}}
\end{equation*}

Since you have used the TF-IDF vectorizer calculating the dot product will directly give you the cosine similarity score. Therefore you will use sklearn's linear_kernel() instead of cosine_similarities() since it is faster.

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Example of the output
print(cosine_sim[:3, :3])
```

Each row of the similarity matrix represents each item in our database. So row 0 = first_item and row 1 = sencond_item. The same thing applies for columns. Due to this the diagonal is always 1, since every item is 100% similar to itself... In the previous example we can see that the similarity between items 1 and 2 is 1.5% and between items 2 and 3 it's 4.68%.
<br><br>

The recommender will receive an item (movie title) as input and it will return a list of the 10 most similar items. For this we need a reverse mapping of movie titles and DataFrame indices.

The steps to follow in the recommender will be the following:

1. Get the index of the movie given its title.
2. Get the list of cosine similarity scores for that particular movie against all movies. Convert it into a list of tuples where the first element is its position and the second is the similarity score.
3. Sort the aforementioned list of tuples based on the similarity scores.
4. Get the top 10 elements of this list. Ignore the first element as it refers to self (the movie most similar to a particular movie is the movie itself).
5. Return the titles corresponding to the indices of the top elements.

```python
# Reverse map of indices and movie titles
indices = pd.Series(metadata.index, index=metadata['title']).drop_duplicates()
```

```python
def get_recommendations(title, cosine_sim=cosine_sim, topn=10):
    """Takes in movie title as input and outputs most similar movies"""
    if isinstance(title, str):
        title = [title]
    sim_scores = list()
    
    for t in title:
        # Get the index of the movie that matches the title
        idx = indices[t]

        # Get the pairwsie similarity scores of all movies with that movie
        mov_list = list(enumerate(cosine_sim[idx]))
        del mov_list[idx]            # Ignore itself
        sim_scores.extend(mov_list)

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies
    sim_scores = sim_scores[:(topn+1)]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the topn most similar movies
    return metadata['title'].iloc[movie_indices]
```

From the new features (cast, crew and keywords) extract the most important actors, the director and the keywords. Right now the data is in the form of "stringified" lists so first of all we transform than into usable lists.

```python
# Example of data...
metadata['cast'][0][:99]
```

```python
# Parse the stringified features into their corresponding python objects
from ast import literal_eval

features = ['cast', 'crew', 'keywords', 'genres']
for feature in features:
    metadata[feature] = metadata[feature].apply(literal_eval)
```

Now we extract the required information from each feature.

```python
def get_director(x):
    """Get the director's name from the crew feature. If director is not listed, return NaN"""
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return np.nan
```

```python
def get_list(x, topn=3):
    """Returns the list top n elements or entire list, whichever is more."""
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than n elements exist. If yes, return only first ones. If no, return entire list.
        if len(names) > topn:
            names = names[:topn]
        return names

    #Return empty list in case of missing/malformed data
    return []
```

```python
# Define new director, cast, genres and keywords features that are in a suitable form.
metadata['director'] = metadata['crew'].apply(get_director)

features = ['cast', 'keywords', 'genres']
for feature in features:
    metadata[feature] = metadata[feature].apply(get_list)
```

We don't want to count as different things in uppercase and lowercase or confusion because of the spaces. This is done so that vectorizer doesn't count the Marlon of "Marlon Brando" and "Marlon Jackson" as the same. After this step the previous actors will be distinct to the vectorizer.

```python
# Function to convert all strings to lower case and strip names of spaces
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i.replace(" ", "")) for i in x]
    else:
        # Check if director exists. If not, return empty string
        if isinstance(x, str):
            return str.lower(x.replace(" ", ""))
        else:
            return ''
```

```python
# Apply clean_data function to your features.
features = ['cast', 'keywords', 'director', 'genres']

for feature in features:
    metadata[feature] = metadata[feature].apply(clean_data)
```

```python
def create_combination(x):
    return ' '.join(x['keywords']) + ' ' + ' '.join(x['cast']) + \
           ' ' + x['director'] + ' ' + ' '.join(x['genres'])
```

### Fastai Collaborative Filtering

FastAI provides two different types of collab models. A simple model called EmbeddingDotBias which used for almost all recommendation systems a few years ago. It creates embeddings for both users and items and then takes the dot product of them. The second is a neural network based model which uses embeddings and fully-connected layers.

An embedding is a mapping from discrete objects, such as words or ids of items and users, to a vector of continuous values. This can be used to find similarities between the discrete objects, that wouldn’t be apparent to the model if it didn’t use embedding layers.
These embedding vectors are low-dimensional and get updated whilst training the network.
Both models can be created using the collab_learner class. Standardly the use_nn argument is set to false and therefore we are creating an EmbeddingDotBias model.

The fastai library contains a *CollabFilteringDataset* class that will help you create datasets suitable for training, and a function *get_colab_learner* to build a simple model directly from a ratings table.

We'll first turn the userId and movieId columns in category codes, so that we can replace them with their codes when it's time to feed them to an Embedding layer. This step would be even more important if our csv had names of users, or names of items in it. To do it, we simply have to call a *CollabDataBunch* factory method.

#### EmbeddingDotBias

```python
# Load data for fastai
data_fastai = CollabDataBunch.from_df(ratings_sm, seed=42, user_name='userId', 
                                      item_name='movieId', rating_name='rating')
```

```python
# Model parameters
n_factors = 50                   # n_factors - dimension of embedding matrix (D)
epochs = 5
learning_rate = 5e-3
wd = 2e-4                        # weight decay

# Create a Learner object. By default it's a EmbeddingDotBias model (use_nn=False)
learn_dot_bias = collab_learner(data_fastai, n_factors=n_factors, y_range=(0.,5.))
# other params: use_nn=True, layers=[50, 50]

learn_dot_bias.lr_find()         # find learning rate
learn_dot_bias.recorder.plot()   # plot learning rate graph
```

```python
# Begin training
learn_dot_bias.fit_one_cycle(epochs, learning_rate, wd=wd)
```

#### EmbeddingNN

```python
learn_nn = collab_learner(data_fastai, use_nn=True, 
                          layers=[256, 128], y_range=(1, 5),
                          emb_szs={'userId': 40, 'movieId':40})
learn_nn.lr_find()           # find learning rate
learn_nn.recorder.plot()     # plot learning rate graph
```

```python
# Begin training
learn_nn.fit_one_cycle(epochs, learning_rate, wd=wd)
```

### Interpreting Embeddings

Because the embeddings learned should represent the **style** and **kind** of **items and users** as best as possible, they might contain interesting features which we can extract and visualize in order of getting an inside view of our data.

Now we will extract the biases and weights of the items in order to get insights of what items should be ranked lower or higher using the embedding bias, as well as how similar some of the most popular itmes are using the embedding weights.

```python
# get top items
g = ratings_sm.groupby('movieId')['rating'].count()
top_movies = g.sort_values(ascending=False).index.values[:1000]
top_movies = top_movies.astype(str)

# create array containing the names of the top movies
top_movies_with_name = []
for movie in top_movies[::-1]:
    rst = metadata[(metadata['id']==int(movie))]
    if rst.shape[0] > 0:
        top_movies_with_name.append(rst['title'].iloc[0])
    else:
        top_movies = np.delete(top_movies, np.where(top_movies == movie))
top_movies_with_name = np.array(top_movies_with_name)
```

Now we can extract the biases for the top movies as well as the mean rating of the top movies. With this, we can get information about the movies that are generally rated low or high no matter what user is rating them.

```python
# get biases for top movies
movie_bias = learn_dot_bias.bias(top_movies, is_item=True)

# get mean ratings
movie_mean_ratings = ratings_sm.groupby('movieId')['rating'].mean()
movie_ratings = [(b, top_movies_with_name[i], movie_mean_ratings.loc[int(tb)]) 
                 for i, (tb, b) in enumerate(zip(top_movies, movie_bias))]

# print movie_bias info
element_0 = lambda e:e[0]
print(f"\nLower ratings:")
for m in sorted(movie_ratings, key=element_0)[:15]:
    print(m)
print(f"\nTop ratings:")
for m in sorted(movie_ratings, key=element_0, reverse=True)[:15]:
    print(m)

# get weights
item_w = learn_dot_bias.weight(top_movies, is_item=True)

# transform weights to 3 dimensions
item_pca = item_w.pca(3)

# get prinicipal components
fac0, fac1, fac2 = item_pca.t()
item_comp0 = [(f, i) for f, i in zip(fac0, top_movies_with_name)]
item_comp1 = [(f, i) for f,i in zip(fac1, top_movies_with_name)]

# print fac0 information
print(f"\nFactor 0 Info:")
for f0 in sorted(item_comp0, key=itemgetter(0), reverse=True)[:10]:
    print(f0)
for f0 in sorted(item_comp0, key=itemgetter(0))[:10]:
    print(f0)

# print fac1 information
print(f"\nFactor 1 Info:")
for f1 in sorted(item_comp1, key=itemgetter(0), reverse=True)[:10]:
    print(f1)
for f1 in sorted(item_comp1, key=itemgetter(0))[:10]:
    print(f1)
```
